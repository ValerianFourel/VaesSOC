```python
import os
import torch

def convert_pt_files_to_float(input_folder, output_folder):
    """
    Converts tensors in .pt files to float and saves them to new .pt files.

    Args:
        input_folder (str): Path to the input folder containing .pt files.
        output_folder (str): Path to the output folder for saving converted .pt files.
    """
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # List all .pt files in the input folder
    pt_files = [f for f in os.listdir(input_folder) if f.endswith(".pt")]

    for pt_file in pt_files:
        input_path = os.path.join(input_folder, pt_file)
        output_path = os.path.join(output_folder, pt_file)

        try:
            # Load the tensor from the input file
            tensor = torch.load(input_path)

            # Convert the tensor elements to float
            tensor_float = tensor.float()

            # Save the new tensor to the output file
            torch.save(tensor_float, output_path)

            print(f"Converted '{input_path}' to float and saved as '{output_path}'")
        except Exception as e:
            print(f"An error occurred while processing '{input_path}': {e}")

```

```python
# Usage example
input_folder = "/content/dataElevation"
output_folder = "/content/dataElevationTensor"
convert_pt_files_to_float(input_folder, output_folder)
```

```python
import os
import shutil

def zip_and_move_folder(source_folder, zip_destination, move_destination):
    """
    Zips a folder and moves the resulting zip file to a specified destination.

    Args:
        source_folder (str): Path to the folder to be zipped.
        zip_destination (str): Path to the location where the zip file will be saved.
        move_destination (str): Path to the destination where the zip file will be moved.
    """
    # Create the destination directory if it doesn't exist
    os.makedirs(zip_destination, exist_ok=True)
    os.makedirs(move_destination, exist_ok=True)

    # Get the base name of the source folder
    folder_name = os.path.basename(source_folder)

    # Create a zip file
    zip_filename = f"{folder_name}.zip"
    zip_path = os.path.join(zip_destination, zip_filename)
    shutil.make_archive(os.path.splitext(zip_path)[0], 'zip', source_folder)

    print(f"Successfully zipped '{source_folder}' to '{zip_path}'")

    # Move the zip file to the move destination
    move_path = os.path.join(move_destination, zip_filename)
    shutil.move(zip_path, move_path)

    print(f"Moved '{zip_path}' to '{move_path}'")
```
